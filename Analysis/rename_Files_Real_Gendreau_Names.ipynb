##### Change filename based on map (txt files)

```python
# Loop through the files in the folder
folder_path = r"Gendreau_et_al_2006"
for filename in os.listdir(folder_path):
    base, ext = os.path.splitext(filename)
    if base in file_map:
        new_name = file_map[base] + ext  # retain the same extension
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} ‚ûù {new_name}")
```

#### Change the first line of the txt files from old names to new names! 

```python

# Folder path
folder_path = r"Gendreau_et_al_2006"
# Loop over all renamed files
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    base, ext = os.path.splitext(filename)
    
    # Only process text files (optional filter)
    if ext.lower() != ".txt":
        continue

    # Read lines from file
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    # Replace first line with new name
    if lines:
        lines[0] = f"Name\t\t\t\t{base}\n"

    # Write back updated lines
    with open(file_path, "w", encoding="utf-8") as file:
        file.writelines(lines)

    print(f"Updated first line in: {filename}")
```
